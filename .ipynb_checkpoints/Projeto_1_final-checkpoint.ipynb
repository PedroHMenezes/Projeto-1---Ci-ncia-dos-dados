## MERCOSUL

Mercosul (Mercado Comum do Sul) integra os territórios da América do Sul em um tratado econômico, fundada pelos países: Brasil, Argentina, Uruguai e Paraguai. Criada em 1991, trouxe mudanças importantes para o continente, tendo em vista que o próprio comércio da América Latina foi estimulado por conta dos benefícios proporcionados pelo acordo.

### Pergunta
Houve algum país que foi mais beneficiado devido à criação do MERCOSUL, ou seja, teve a possibilidade de aumentar suas exportações ao fazer do grupo?

### Hipótese
Após criar o MERCOSUL, os países que criaram esse bloco econômico estavam em situações economicamente complicadas e por conta disso, usaram as possibilidades que o bloco proporcionava de exportar seus produtos para assim, melhorar seu estado financeiro. Dessa forma, analisando os países que formaram o bloco(Brasil, Argentina, Paraguai e Uruguai), a Argentina era o país que estava mais apto para aproveitar das oportunidades que o bloco oferecia por ter mais estabilidade política e medidas econômicas que visavam a estabilidade financeira.

https://www.imf.org/external/pubs/ft/fandd/2000/03/pou.htm - Este artigo oferece um overview do processo de reforma econômica, denominado plano de conversibilidade, que a Argentina sofreu no início da década de 1990, o que pode ter contribuido para seu crescimento na década que estava começando. 

https://www.worldfinance.com/special-reports/a-history-of-economic-trouble-in-argentina - Esta notícia fornece um aspecto importante do crescimento da Argentina na década de 1990 e posteriormente, nos anos de 2000 para frente, quando ela teve um crescimento de seu GDP per capita sempre positivo

http://countrystudies.us/uruguay/44.htm - Esse artigo apresenta informações sobre o crescimento do Uruguai na década de 1990 após sair do regime militar em que vivia. Um ponto importante desse processo foram as várias dívidas externas que o país tinha e que de certo modo impediram um crescimento maior da economia.

https://data.worldbank.org/indicator/NY.GDP.MKTP.KD.ZG?locations=UY - Esse gráfico é sobre o crescimento do GDP do Uruguai desde 1961, com destaque para os anos de 1990 em diante, nos quais esse crescimento foi sempre positivo.

https://pt.wikipedia.org/wiki/Paraguai - Este artigo mostra sobre a instabilidade do Paraguai no começo da década de 90, com deposição de presidentes em 1989. Com este quadro de instabilidade, o crescimento da economia foi bem debilitado.

https://books.google.com.br/books?id=1oCRcHJcflcC&pg=PA5&lpg=PA5&dq=paraguay+economcs+1990&source=bl&ots=1qqQLQy4QA&sig=ACfU3U29SoO67hibkRuhqXI9QdPArYTjFA&hl=pt-BR&sa=X&ved=2ahUKEwib3fe-v57kAhUgGLkGHSE0DJEQ6AEwBnoECAgQAQ#v=onepage&q&f=false - O overview no início deste artigo permite entender que as condições financeiras do Paraguai estavam prejudicadas, sobretudo por conta do quadro político.

https://en.wikipedia.org/wiki/Economic_history_of_Brazil#1990%E2%80%9393_period - Neste é apresentado a situação do Brasil na década de 1990, na qual passou por situações complicadas economicamente até a criação do plano real em 1994 que conseguiu diminuir a inflação alta que estava na economia.

### Código implementado
Acessando a tabela de GDP per capita:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import os
from numpy import arange
from numpy import percentile
import numpy as np
import matplotlib
#matplotlib.style.use()
```

```python
gdp = pd.read_excel('GDPpercapitaconstant2000US.xlsx')
```

Acessando a tabela regiões:

```python
region = pd.read_excel('landlocked_region.xlsx')
```

Acessando a tabela de exportações:

```python
exportation=pd.read_excel("Exports (p of GDP).xlsx")
exportation_index = exportation.loc[:,['Exports of goods and services (% of GDP)','1990']]
exportation_novo = exportation_index.set_index('Exports of goods and services (% of GDP)')
```

Para início da análise, vamos estudar como o GDP per capita dos países estava em 1990, um ano antes da MERCOSUL ser criada

Formatando as tabelas de região e gdp para uni-las.

```python
gdp_1990 = gdp.loc[:,['Income per person (fixed 2000 US$)', "1990"]]
gdp_index = gdp_1990.set_index('Income per person (fixed 2000 US$)')
```

```python
region_index = region.set_index('Country')
```

União das tabelas:

```python
exp_r=gdp_index.join(exportation_novo,how="inner",lsuffix='_gdp',rsuffix='_exp')
exp_region = exp_r.join(region_index,how='inner')
```

Unida as tabelas, vamos analisar como estava a condição na América do Sul em 1990

### Antes da criação do bloco
Como estavam os países da América Latina antes da criação do MERCOSUL, com destaque para a Argentina, Brasil, Uruguai e Paraguai.

```python
mercosul=[1,0,1,0,0,0,0,1,0,0,1,0]
exp_south_america = exp_region.loc[(exp_region['region']=='america_south')]
exp_south_america['Mercosul']=mercosul
```

Analisando a porcentagem de exportação dos países que **não fazem** parte do MERCOSUL

```python
exp_south_america.loc[(exp_south_america['Mercosul'] == 0)].boxplot(column='1990_exp')
```

Analisando a porcentagem de exportação dos países que **fazem** parte do MERCOSUL

```python
exp_south_america.loc[(exp_south_america['Mercosul'] == 1)].boxplot(column='1990_exp')
```

É possível observar a partir dos gráficos acima que o nível de exportação dos países fora do bloco analisado é bem maior que os de dentro, ou seja, a exportação nesses países não é a principal fonte de renda.   

Vamos analisar então o GDP per capita dos países da América do Sul para ver se algum dos que compõe o bloco se destaca.

```python
exp_south_america = exp_region.loc[(exp_region['region']=='america_south')]
exp_south_america['Mercosul']=mercosul
exp_south_america.head()
```

Com a tabela acima e possível observar que a Argentina se destaca como o país que tinha o GDP per capita mais alto. Dessa maneira, é possível que ela tenha criado o bloco junto aos outros países e ter aproveitado dessa situação econômica mais favorável para exportar seus produtos e crescer financeiramente.

### Depois da criação do bloco
Feita a análise prévia antes da criação do MERCOSUL, vamos estudar o comportamento do GDP per capita durante os anos que se seguiram. Em primeiro lugar, é necessário estudar o comportamento dos países que faziam parte desse bloco.

```python
gdp_indexado = gdp.set_index('Income per person (fixed 2000 US$)')
gdp_paises = gdp_indexado.T
gdp_america = gdp_paises.loc[:,['Argentina','Bolivia','Brazil','Chile','Colombia','Ecuador','Guyana','Paraguay','Peru','Suriname','Uruguay','Venezuela']]
gdp_paises_index = gdp_america.T
gdp_paises_index = gdp_paises_index.loc[:,['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001']]
gdp_1990_paises = gdp_paises_index.T
gdp_1990_paises
anos = ['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001']
anos_plt = [0]*12
i=0
for k in gdp_1990_paises:
    if k=='Argentina' or k=='Brazil' or k=='Uruguay' or k=='Paraguay':
        plt.plot(anos,gdp_1990_paises[k],label='{}'.format(k))
    i+=1
plt.xlabel('Anos')
plt.ylabel('GDP per capita(fixado 2000 US$)')
plt.title('GDP per capita de 1990 a 2001 dos países do MERCOSUL')
plt.grid(True)
plt.legend()
plt.show()
```

Uma primeira conclusão possível de ser tirada é a de que a Argentina teve um crescimento bem grande durante os 10 primeiros anos do bloco, sendo o país com maior GDP per capita do grupo tanto no momento que foi criado quanto 10 anos depois.

Outro ponto interessante de se analisar é o progresso do GDP per capita da América do Sul como um todo. Dessa forma, o gráfico a seguir demonstra como é esse comportamento desse indicador ao longo do tempo.

```python
for k in gdp_1990_paises:
    #for l in range(0,len(anos_plt)):
        #anos_plt[l] = anos[i]
    plt.plot(anos,gdp_1990_paises[k],label='{}'.format(k))
    i+=1
plt.xlabel('Anos')
plt.ylabel('GDP per capita(fixado 2000 US$)')
plt.title('GDP per capita de 1990 a 2001 nos países da América do Sul')
plt.grid(True)
plt.legend()
plt.show()
```

Com o gráfico acima, é possível analisar que de fato a Argentina tem o maior GDP per capita do continente durante todo o período analisado, reforçando que ela pode ter sido a mais beneficiada do bloco. Outro ponto a ser levantado é o crescimento relativo de 1990 a 2001. Vamos construir essa análise a seguir.

```python
gdp_paises = gdp_indexado.T
gdp_america = gdp_paises.loc[:,['Argentina','Bolivia','Brazil','Chile','Colombia','Ecuador','Guyana','Paraguay','Peru','Suriname','Uruguay','Venezuela']]
gdp_paises_index = gdp_america.T
gdp_paises_index = gdp_paises_index.loc[:,anos]
gdp_1990_paises = gdp_paises_index.T
gdp_1990_paises
decimal_gdp = gdp_1990_paises.T['2001']/gdp_1990_paises.T['1990']
decimal_gdp
```

Utilizando os valores acima de crescimento relativo do GDP per capita, criou-se um gráfico para melhor visualização dos dados e assim tirar conclusões a partir dele

```python
paises_plt = ['Arg','Bol','Bra','Chi','Col','Ecu','Guy','Par','Per','Sur','Uru','Ven']
perc_gdp = []
for j in decimal_gdp:
    perc_gdp.append(j*100)
plt.plot(paises_plt,perc_gdp,'ro')
plt.title('Taxa de crescimento do GDP per capita dos Países da América do Sul (%)')
plt.xlabel('Países')
plt.ylabel('Taxa de crescimento GDP per capita')
plt.grid(True)
plt.show()
```

A partir dessa análise, é possível dizer que em percentual, houve dois outros países na américa do sul que o GDP per capita cresceu mais que a Argentina (Chile e Guyana).

É perceptível que houve crecumento no GDP per capita da Argentina, porém ainda falta analisar se esse crescimento se refletiu em crescimento das exportações ou não. Vamos analisar isso agora.

```python
exp_indexado = exportation.set_index('Exports of goods and services (% of GDP)')
exp_paises = exp_indexado.T
exp_america = exp_paises.loc[:,['Argentina','Bolivia','Brazil','Chile','Colombia','Ecuador','Guyana','Paraguay','Peru','Suriname','Uruguay','Venezuela']]
exp_paises_index = exp_america.T
exp_paises_index = exp_paises_index.loc[:,['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001']]
exp_1990_paises = exp_paises_index.T
exp_1990_paises
anos = ['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001']
anos_plt = [0]*12
i=0
for k in exp_1990_paises:
    if k=='Argentina' or k=='Brazil' or k=='Uruguay' or k=='Paraguay':
        plt.plot(anos,exp_1990_paises[k],label='{}'.format(k))
    i+=1
plt.xlabel('Anos')
plt.ylabel('Exports of goods and services (% of GDP)')
plt.title('Taxa de exportação em % do GDP para os países do MERCOSUL de 1990 a 2001')
plt.grid(True)
plt.legend()
plt.show()
```

A partir desse gráfico é possível entender que o crescimento do GDP per capita da Argentina não foi ocasionado por conta do aumento da exportações, pois a participação desse segmento na economia argentina não mudou ao longo dos anos, ou seja, a Argentina não foi a maior beneficiada pela criação no MERCOSUL no sentido de aproveitar o mercado consumidor que o bloco proporcionou para vender seus produtos.

## Conclusão
Respondendo à pergunta e analisando como o crescimento dos países dentro do MERCOSUL e da América Latina aconteceu, é possível dizer que a Argentina foi o país que teve maior desenvolvimento relativo dentro do bloco econômico(130%, contra 110% do Brasil, 94% do Paraguai e 126% do Uruguai). Contudo, dentro da América Latina, não foi o país que teve destaque como maior crescimento relativo do GDP per capita, atrás da Guyana(162%) e do Chile(171%).
Entretanto, o crescimento do seu GDP per capita não refletiu no aumento das exportações como parte significativa do GDP, ou seja, esse desenvolvimento econômico se deu por outros meios que não a exportação de seus produtos para o bloco econômico que faz parte.
